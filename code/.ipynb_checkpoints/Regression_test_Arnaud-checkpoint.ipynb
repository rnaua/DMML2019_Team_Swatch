```python
def ln_regression(valX,valy,labelX,labelY,title):
    x = valX
    y = valy
    
    r = LinearRegression()
    r.fit(x,y)
    p = r.predict(x)
    
    mse = mean_squared_error(y,p)
    plt.title("%s\nDegree 1\n MSE: %.2f" % (title, mse))
    plt.scatter(x.iloc[:,0],y)
    plt.plot(x.iloc[:,0], p, 'r')
    plt.xlabel(labelX)
    plt.ylabel(labelY)
    plt.show()
    return

def regression(valX,valY,deg,labelX,labelY,title):
    if deg in [0,1]:
        return ln_regression(valX,valY,labelX,labelY,title)
    poly_reg = PolynomialFeatures(degree=deg)
    x = poly_reg.fit_transform(valX)
    y = valY
    lr = LinearRegression()
    lr.fit(x,y)
    p = lr.predict(x)
    
    rmse = np.sqrt(mean_squared_error(y,p))
    plt.title("%s\nDegree %d\n MSE: %.2f" % (title,deg, rmse))
    plt.scatter(valX, valY, color='orange', label="samples")
    plt.plot(valX, p, color='blue')
    plt.xlabel(labelX)
    plt.ylabel(labelY)
    plt.show()
    return


```

# Forecasting

## Forecasting with AR,MA and ARIMA

```python
test = persons.groupby("Date")[['Value']].sum().reset_index()
ts = test[['Date','Value']]
ts['Date'] = pd.to_datetime(test['Date'])
ts.set_index('Date',inplace=True)
ts = ts['Value']
```

```python

plt.figure(figsize=(20, 10))
plt.plot(ts)
```

The plot shows that there is an overall decrease over the years, with some seasonality in it.

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()
#Plot rolling statistics:
    plt.figure(figsize=(20, 10))
    plt.plot(timeseries, color='blue',label='Original')
    plt.plot(rolmean, color='red', label='Rolling Mean')
    plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
    
test_stationarity(ts)
```

```python
ts_log = np.log(ts)
ts_log_diff = ts_log - ts_log.shift()
plt.figure(figsize=(20, 7))
plt.plot(ts_log_diff)
```

```python
ts_log_diff.dropna(inplace=True)
test_stationarity(ts_log_diff)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(ts_log)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.seasonal

plt.figure(figsize=(30, 15))
plt.subplot(411)
plt.plot(ts_log,label = 'Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend,label = 'Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label = 'Seasonality')
plt.legend(loc='best')
```

```python
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(ts_log,order=(2,1,0))
results_AR = model.fit(disp=-1)
plt.figure(figsize=(30, 15))
plt.plot(ts_log_diff)
plt.plot(results_AR.fittedvalues,color='red')
plt.title('RSS : %.4f'% sum((results_AR.fittedvalues-ts_log_diff)**2))
```

```python
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(ts_log,order=(0,1,2))
results_MA = model.fit(disp=-1)
plt.figure(figsize=(30, 15))
plt.plot(ts_log_diff)
plt.plot(results_MA.fittedvalues,color='red')
plt.title('RSS : %.4f'% sum((results_MA.fittedvalues-ts_log_diff)**2))
```

```python
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(ts_log,order=(2,1,2))
results_ARIMA = model.fit(disp=-1)
plt.figure(figsize=(30, 15))
plt.plot(ts_log_diff)
plt.plot(results_ARIMA.fittedvalues,color='red')
plt.title('RSS : %.4f'% sum((results_ARIMA.fittedvalues-ts_log_diff)**2))
```

```python
prediction_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy =True)
print (prediction_ARIMA_diff.head())
```
